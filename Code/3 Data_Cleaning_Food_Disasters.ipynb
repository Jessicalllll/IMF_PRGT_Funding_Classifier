```python
import pandas as pd
```

To preprocess the natural disaster data, we imported the EM-DAT data as a pandas DataFrame. Then, we used the groupby operation and the count aggregate function to calculate the number of natural disasters experienced by each country by year.

To preprocess the food supply data, we imported the FAOSTAT data as a pandas DataFrame. Because the database used different methodologies before and after 2010, the data was divided into two parts that have the same structure. We first merged the two datasets vertically, then removed all columns except for “Area”, “Year”, and “Value”. Lastly, we renamed the “Value” column as “Food Supply (kcal/capita/day)”.


# Natural Disasters

```python
nd_df = pd.read_excel('Natural_Disasters_Copy.xlsx')
nd_df
```

```python
nd_final_df = nd_df.groupby(['Country', 'Year'])[['Dis No']].count().reset_index()
nd_final_df.rename(columns={'Dis No':'Number of Disasters'}, inplace=True)
nd_final_df
```

```python
nd_final_df.to_csv('Natural_Disasters_Cleaned')
```

# Food Balance

```python
fb_df = pd.read_csv('Food_Supply.csv', encoding='latin-1')
fb_df
```

```python
fb_final_df = fb_df[['Area', 'Year', 'Value']]
fb_final_df.rename(columns={'Value':'Food Supply (kcal/capita/day)'}, inplace=True)
fb_final_df
```

```python
fb_final_df.to_csv('Food_Supply_Cleaned')
```
